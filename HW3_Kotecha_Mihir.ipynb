### Exercise 1


Suppose X has density $f_{x}(s) = 2s\textbf{1}\{s \in [0,1] \}$.

**Explain in video.** Generate 1000 random variables from X's distribution. Remember to show any math needed to do this. Then plot them in a histogram and add a line of the target density for comparison.

Hint: if you set the seed to 1, the histogram should look like this (you need to also add the line though):


```r
# Setting the number of random variables to generate
n_samples <- 1000

# Generating the random values from a uniform distribution
uniform_rv <- runif(n_samples, 0, 1)

# The CDF of the given density function is x^2, and the inverse CDF is sqrt(y); apply the inverse CDF to the uniform random variables to transform them
generated_rv <- sqrt(uniform_rv)

# Plotting the histogram of the generated RVs
hist(generated_rv, probability = TRUE, main = "Histogram of Random Variables from Target Distribution",
     xlab = "Generated Random Variables", ylim = c(0, 2))

# Creating a sequence for plotting the target density; adding a line for the target density function
x_vals <- seq(0, 1, length.out = 100)
lines(x_vals, 2 * x_vals, col = "red")
```

### Exercise 4

A discrete random variable $X$ has probability mass function

| $x$ | 0 | 1 | 2 | 3 | 4 |
| -------- | -------- | -------- | -------- | -------- | -------- |
| $p(x)$   | 0.1  | 0.2    | 0.2    | 0.2    | 0.3   |

a)  **Explain in Video:** Use the inverse transform method to generate a random sample of size 1000 from the distribution of $X$. Construct a relative frequency table. Also make a bar plot.

Hint: if you set the seed to 1, the first 3 values of X you get should be `1, 2, 3` (I know that's a weird coincidence, but that's really the order of the first three rvs).

b) **Explain in Video:** Now instead use R's built in `sample` function to generate a random sample of size 1000 from the distribution of $X$. Construct a relative frequency table. Also make a bar plot.

Hint: if you set the seed to 1, the first 3 values of X you get should be `4, 2, 3`.

c) **Write only** Compare the empirical (the relative frequencies you got from the sample) with the theoretical probabilities and with the relative frequencies you get from using the `sample` function.


```r
# Question 4A - Inverse Transform Method
# Set seed for reproducibility
set.seed(1)

# Define the sample size
sample_size <- 1000

# Define the probability mass function for the given values of X
pmf_values <- c(0.1, 0.2, 0.2, 0.2, 0.3)

# Generate uniform random variables between 0 and 1
uniform_samples <- runif(sample_size)

# Create a vector to store the generated X values
generated_values <- numeric(sample_size)

# Use the inverse transform method to assign values to the generated X values
for (i in 1:sample_size) {
  if (uniform_samples[i] <= 0.1) {
    generated_values[i] <- 0
  } else if (uniform_samples[i] <= 0.3) {
    generated_values[i] <- 1
  } else if (uniform_samples[i] <= 0.5) {
    generated_values[i] <- 2
  } else if (uniform_samples[i] <= 0.7) {
    generated_values[i] <- 3
  } else {
    generated_values[i] <- 4
  }
}

# Create a relative frequency table for the generated values
relative_freq <- table(generated_values) / sample_size

# Create a data frame for comparison of empirical and theoretical probabilities
results_table <- data.frame(Empirical = relative_freq, Theoretical = pmf_values,
                            row.names = c("x = 0", "x = 1", "x = 2", "x = 3", "x = 4"))

# Display the results table
results_table

# Create a bar plot for the relative frequencies
barplot(relative_freq, main = "Bar Plot of Generated X Values",
        xlab = "Values of X", ylab = "Relative Frequency", col = "lightblue")
```

```r
# Question 4B - Built-in Sample function
set.seed(1)

# Defining the sample size
sample_size <- 1000

# Defining the values of X and the probabilities
x_values <- 0:4
x_probabilities <- c(0.1, 0.2, 0.2, 0.2, 0.3)

# Using the sample function to generate random values from the distribution of X
sampled_values <- sample(x_values, size = sample_size, replace = TRUE, prob = x_probabilities)

# Creating a relative frequency table for the sampled values
relative_freq_sample <- table(sampled_values) / sample_size

# Creating a data frame to compare empirical and theoretical probabilities
comparison_table <- data.frame(Empirical = relative_freq_sample, Theoretical = x_probabilities,
                               row.names = paste("x =", x_values))

comparison_table

# Creating a bar plot of the relative frequencies
barplot(relative_freq_sample, main = "Bar Plot of Sampled X Values Using sample()",
        xlab = "Values of X", ylab = "Relative Frequency", col = "lightgreen")
```

```r
# Question 4C - Comparing emperical and theoretical frequencies
comparison_results <- data.frame(
  Empirical_Transform = table(generated_values) / sample_size,  # From inverse transform method
  Empirical_Sample = table(sampled_values) / sample_size,       # From sample() function
  Theoretical = x_probabilities,                                # Theoretical probabilities
  row.names = paste("x =", x_values)
)

# Clean up the data frame by removing redundant columns (if necessary)
comparison_results <- subset(comparison_results)

# Display the comparison of results
comparison_results
```

From the comparison table, we can see that both the empirical frequencies generated using the inverse transform method and the sample function closely match the theoretical probabilities for each value of X. The slight differences between the empirical and theoretical values are due to the randomness in the sampling process.
